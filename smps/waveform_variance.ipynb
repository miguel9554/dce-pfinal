```python
import csv
from matplotlib import pyplot as plt
import numpy as np

current = []
time = []
with open('simulaciones/fuente.txt', 'r') as csvfile:
    reader = csv.DictReader(csvfile, delimiter='\t')
    for row in reader:
        current.append(float(row['-I(L1)']))
        time.append(float(row['time']))

current = np.array(current)
time = np.array(time)
```

```python
plt.plot(time*1e3, current)
plt.xlabel('Time [ms]')
plt.ylabel('Current [A]')
plt.grid()
```

```python
delta_I = 2*np.sqrt(np.var(
    current[np.where(time > 11e-3)]
))

print(delta_I)
```
