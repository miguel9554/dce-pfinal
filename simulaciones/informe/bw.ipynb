```python
import csv
import re
f_sim_10w = []
mag_sim_10w = []
phase_sim_10w = []
with open('bw_10w.txt', encoding = 'unicode_escape') as csvfile:
    spamreader = csv.reader(csvfile, delimiter='\t')
    first = True
    for row in spamreader:
        if first:
            first = False
            continue
        f_sim_10w.append(float(row[0]))
        p = re.compile(r'\((?P<magnitude>.*)dB,(?P<phase>.*)°\)')
        m = p.search(row[1])
        mag_sim_10w.append(float(m.group('magnitude')))
        phase_sim_10w.append(float(m.group('phase')))
```

```python
f_sim_OL = []
mag_sim_OL = []
phase_sim_OL = []
with open('bw_OL.txt', encoding = 'unicode_escape') as csvfile:
    spamreader = csv.reader(csvfile, delimiter='\t')
    first = True
    for row in spamreader:
        if first:
            first = False
            continue
        f_sim_OL.append(float(row[0]))
        p = re.compile(r'\((?P<magnitude>.*)dB,(?P<phase>.*)°\)')
        m = p.search(row[1])
        mag_sim_OL.append(float(m.group('magnitude')))
        phase_sim_OL.append(float(m.group('phase')))
```

```python
plt.figure(figsize=(8, 6))
plt.semilogx(f_sim_10w, mag_sim_10w, label='simulación 10 W')    # Bode magnitude plot
plt.semilogx(f_sim_OL, mag_sim_OL, label='simulación OL')    # Bode magnitude plot
plt.semilogx(f_medicion_10W, mag_medicion_10W, label='medición 10 W')    # Bode magnitude plot
plt.semilogx(f_medicion_OL, mag_medicion_OL, label='medición OL')    # Bode magnitude plot
plt.grid()
plt.title('Magnitud de la respuesta en frecuencia')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Ganancia [dB]')
plt.legend()
plt.savefig('bw_gain.png')
```
